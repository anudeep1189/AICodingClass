```python
# 4

def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False
    
input_list = [12,20,77,9,11,1]

for item in input_list:
    isEven = is_even(item)
    if isEven:
        print(f"Number {item} is even")
    else:
        print(f"Number {item} is odd")

```

```python
# 5 
def student_result(marks):
    if marks >= 40:
        return "pass"
    else:
        return "fail"

studentsMarks_List = {"abc":20,"lkj":30,"erd":90,"fgt":40}

for name,marks in studentsMarks_List.items():
    result = student_result(marks)
    print(f"{name} is {result}")

```

```python
# 8 

def get_student_grade(marks):
    if marks >= 90:
        print("A")
    elif marks >=75:
        print("B")
    elif marks >= 60:
        print("C")
    elif marks >= 40:
        print("D")
    else:
        print("Fail")

student_marks=[12,23,45,23,80,76,54,67,43,22]

for item in student_marks:
    get_student_grade(item)
```

```python
def total_marks(*marks):
    total = 0

    for mark in marks:
        total = total + mark

    return total


result = total_marks(80, 70, 90, 60)
print(result)
```

```python
# 10
def average_marks(*marks):
    if len(marks) == 0:
        return "Please provide at least one mark"

    average = sum(marks) / len(marks)
    return average


print(average_marks(80, 70, 90, 60))  
print(average_marks())
```
