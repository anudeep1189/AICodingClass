```python
# 4

def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False
    
input_list = [12,20,77,9,11,1]

for item in input_list:
    isEven = is_even(item)
    if isEven:
        print(f"Number {item} is even")
    else:
        print(f"Number {item} is odd")

```

```python
# 8 

def get_student_grade(marks):
    if marks >= 90:
        print("A")
    elif marks >=75:
        print("B")
    elif marks >= 60:
        print("C")
    elif marks >= 40:
        print("D")
    else:
        print("Fail")

student_marks=[12,23,45,23,80,76,54,67,43,22]

for item in student_marks:
    get_student_grade(item)
```

```python
def total_marks(*marks):
    total = 0

    for mark in marks:
        total = total + mark

    return total


result = total_marks(80, 70, 90, 60)
print(result)
```

```python
# 10
def average_marks(*marks):
    if len(marks) == 0:
        return "Please provide at least one mark"

    average = sum(marks) / len(marks)
    return average


print(average_marks(80, 70, 90, 60))  
print(average_marks())
```

```python
# 12
def multiply_all(*numbers):
    result = 1
    for number in numbers:
        result *= number
    return result    

print(f"Multiplical of all number = {multiply_all(2,3,4)}")
```

```python
# 13
def department_salary_average(*salaries):
    average = sum(salaries)/len(salaries)
    return average

average = department_salary_average(30000, 40000, 50000, 10000, 1000, 70000)

print(f"Average salary = {average}")
```

```python
# 14

def count_even_odd(*numbers):
    countEven = 0 
    countOdd = 0 
    for number in numbers:
        if number % 2 == 0:
            countEven += 1
        else:
            countOdd += 1

    return countEven,countOdd


countEven,countOdd = count_even_odd(1,2,53,49,25,16,7,5,10,12,23,44,56,9,32)
print(f"Even : {countEven}, Odd : {countOdd}")
    
```

```python
#19
def divide(numerator, denominator, /):
    return numerator / denominator

print(divide(10, 2))
print(divide(numerator=10, denominator=2))
```

```python
#20
def ratio(numerator, denominator, /, *, as_percent=False):
    result = numerator / denominator

    if as_percent:
        return result * 100

    return result


print(ratio(1, 4))

print(ratio(1, 4, as_percent=True))
```
