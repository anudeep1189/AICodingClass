```python
# 4

def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False
    
input_list = [12,20,77,9,11,1]

for item in input_list:
    isEven = is_even(item)
    if isEven:
        print(f"Number {item} is even")
    else:
        print(f"Number {item} is odd")

```

```python
# 10
def average_marks(*marks):
    if len(marks) == 0:
        return "Please provide at least one mark"

    average = sum(marks) / len(marks)
    return average


print(average_marks(80, 70, 90, 60))  
print(average_marks())
```

```python
# 14

def count_even_odd(*numbers):
    countEven = 0 
    countOdd = 0 
    for number in numbers:
        if number % 2 == 0:
            countEven += 1
        else:
            countOdd += 1

    return countEven,countOdd


countEven,countOdd = count_even_odd(1,2,53,49,25,16,7,5,10,12,23,44,56,9,32)
print(f"Even : {countEven}, Odd : {countOdd}")
    
```

```python
#20
def ratio(numerator, denominator, /, *, as_percent=False):
    result = numerator / denominator

    if as_percent:
        return result * 100

    return result


print(ratio(1, 4))

print(ratio(1, 4, as_percent=True))
```

```python
#22
def apply_discount(price, discount=10):
    finalPrice = (price - (price * discount) / 100)
    return finalPrice

finalPrice = apply_discount(1000,discount=40)
print(f"Total = {finalPrice}")    
```

```python
#23
def calculate_final_price(price, tax_rate=0.05, discount=0):
    discountedPrice = price - (price * discount)/100
    afterTaxPrice = discountedPrice - (discountedPrice * tax_rate)
    return afterTaxPrice



finalPrice = calculate_final_price(1000,tax_rate=0.05, discount=30)
print(f"Total = {finalPrice}")
```

```python
#25
def append_bad(value, bucket=[]):
    bucket.append(value)
    return bucket

print(f"append bad = {append_bad(1)}")  
print(f"append bad = {append_bad(2)}")  
print(f"append bad = {append_bad(3)}")  


def append_good(value, bucket=None):
    if bucket is None:
        bucket = []

    bucket.append(value)
    return bucket


print(f"append good = {append_good(1)}")  
print(f"append good = {append_good(2)}")  
print(f"append good = {append_good(3)}")  
```

```python
#26
def add_student_bad(student, student_list=[]):
    student_list.append(student)
    return student_list


print("bad:", add_student_bad("abc"))
print("bad:", add_student_bad("xyz"))
print("bad:", add_student_bad("pqr"))


def add_student_good(student, student_list=None):
    if student_list is None:
        student_list = []

    student_list.append(student)
    return student_list


print("good:", add_student_good("abc"))
print("good:", add_student_good("xyz"))
print("good:", add_student_good("pqr"))
```

```python
#29
def average(*numbers: float) -> float:
    if len(numbers) == 0:
        return 0.0
    return sum(numbers) / len(numbers)

print(average(10, 20, 30))
print(average(5.5, 4.5))
print(average())
```
