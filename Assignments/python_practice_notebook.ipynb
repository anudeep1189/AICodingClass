## Q24  Palindrome Check

**Hint:** Ignore spaces and case while checking.

```python
text = "Never odd or even"
# Check whether text is a palindrome (ignore spaces/case)

# Your solution below:
cleaned = ""
for char in text:
    if char != " ":
        cleaned += char.lower()

if cleaned == cleaned[::-1]:
    print(f'"{text}" is a palindrome')
else:
    print(f'"{text}" is not a palindrome')
```

## Q27  Longest Unique Substring

**Hint:** Sliding window works well for this problem.

```python
text = "abcabcbb"
# Find the length of the longest substring without repeating characters

# Your solution below:
max_len = 0
for i in range(len(text)):
    seen = []
    for j in range(i, len(text)):
        if text[j] in seen:
            break
        seen.append(text[j])
        if len(seen) > max_len:
            max_len = len(seen)

print(f"Longest unique substring length: {max_len}")
```

## Q28  Character Compression

**Hint:** Build a compressed string with counts of consecutive characters.

```python
text = "aaabbcdddd"
# Compress to format like a3b2c1d4

# Your solution below:
compressed = ""
i = 0
while i < len(text):
    char = text[i]
    count = 1
    while i + count < len(text) and text[i + count] == char:
        count += 1
    compressed += char + str(count)
    i += count

print(compressed)
```

## Q40  Intersection Logic Puzzle

## Q45  Merging Nested Dicts: Shallow vs Deep

**Hint:** `dict.update` performs a shallow merge — show consequences with nested dicts and how to perform a deep merge.

```python
a = {'x': {'n': 1}}
b = {'x': {'m': 2}}
res = a.copy()
res.update(b)
# Explain what `res['x']` contains and why; suggest a deep merge approach

# Your solution below:
print(f"res['x'] = {res['x']}")
def deep_merge(base, override):
    result = {}
    for k in base:
        result[k] = base[k]
    for k in override:
        if k in result and type(result[k]) == dict and type(override[k]) == dict:
            result[k] = deep_merge(result[k], override[k])
        else:
            result[k] = override[k]
    return result

deep_res = deep_merge(a, b)
print(f"Deep merge result: {deep_res}")
```
