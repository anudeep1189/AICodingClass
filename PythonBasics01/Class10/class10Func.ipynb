# recurssion function

```python
def coutdown(n):
    if n==0:
        print("over")
        return
    print(n)
    coutdown(n-1)

coutdown(10)    
```

```python
def factorial(n):
    if n==0:
        return 1 
    #print(f"{n} * {factorial(n-1)} ")
    return n * factorial(n-1)

factorial(5)
```

```python
def sum(n):
    if n == 1:      
        return 1
    return n + sum(n-1)

sum(10)
```

```python
def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1  
    else:
        return fib(n - 1) + fib(n - 2)

fib(10)
```
