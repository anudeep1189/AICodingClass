```python
# # question - 2
# attendance = [
#     "Amit", "Neha", "Amit", "Ravi", "Neha",
#     "Amit", "Sara", "Ravi", "Amit"
# ]

# # How many times each student attended ?
# # Find out the Top 2 most regular students in the calss ?    give me answer in chat 



from collections import Counter

attendance = [
    "Amit", "Neha", "Amit", "Ravi", "Neha",
    "Amit", "Sara", "Ravi", "Amit"
]

counts = Counter(attendance)
print(counts)
print(counts.most_common(2))
```

```python
# students = [
#     ("Shreya", "Python"),
#     ("Abhishek","Data Science"),
#     ("Sagar","Python"),
#     ("Nishtha","Python"),
#     ("Rusrith","AI")
# ]
# # write a program to group students based on course they are enrolled into

students = [
    ("Shreya", "Python"),
    ("Abhishek", "Data Science"),
    ("Sagar", "Python"),
    ("Nishtha", "Python"),
    ("Rusrith", "AI")
]

grouped_students = {}

for name, course in students:
    if course not in grouped_students:
        grouped_students[course] = []
    grouped_students[course].append(name)

print(grouped_students)
```
