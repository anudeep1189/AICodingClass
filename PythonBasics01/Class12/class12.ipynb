```python
#chain map
from collections import ChainMap

default_prices = {
    "course_fee": 10000,
    "discount": 0
}

festival_offer = {
    "discount": 2000
}

student_specific = {
    "course_fee": 8000
}

prices = ChainMap(student_specific, festival_offer, default_prices)

print("Course Fee:", prices["course_fee"])
print("Discount:", prices["discount"])

final_price = prices["course_fee"] - prices["discount"]
print("Final Price:", final_price)

print("\nHierarchy of access:")
print("1. student_specific:", student_specific)
print("2. festival_offer:", festival_offer)
print("3. default_prices:", default_prices)
```

```python
class PositiveList(list):
    def append(self, x):
        if x <= 0: raise ValueError("Only positive")
        super().append(x)

a = PositiveList()
a.append(5)
a.append(10)
print(a)
```
