# Python OOP Core Concepts: Quick Reference Summary

This guide outlines the structural elements of Python Object-Oriented Programming (OOP) and maps them directly to their .NET/C# equivalents.

---

- **1. Class**
* [cite_start]**Definition:** The blueprint or template used to construct concrete object instances[cite: 106].
* [cite_start]**C# Analogy:** Directly maps to a standard C# `class`[cite: 107]. [cite_start]It defines the structure, fields, and behavioral patterns for the objects instantiated from it[cite: 108].

---

- **2. Self**
* [cite_start]**Definition:** A compulsory parameter passed as the first argument to any instance method in Python[cite: 109]. [cite_start]It acts as an explicit pointer connecting data and operations to that exact instance[cite: 110].
* [cite_start]**C# Analogy:** The exact equivalent of the `this` keyword[cite: 111].
* [cite_start]**Key Difference:** While C# implicitly hides `this` within class method blocks, Python requires you to explicitly declare `self` in the method signature (e.g., `def __init__(self, courseName):`)[cite: 112]. [cite_start]However, Python passes it automatically behind the scenes when calling the method[cite: 113].

---

- **3. Instance Variable**
* [cite_start]**Definition:** Variables bound to a specific object instance using `self` (e.g., `self.courseName = courseName`)[cite: 114]. [cite_start]They maintain separate, distinct data for different objects created from the same blueprint[cite: 115].
* [cite_start]**C# Analogy:** Non-static class fields or properties (e.g., `public string CourseName { get; set; }`)[cite: 116].

---

- **4. Object**
* [cite_start]**Definition:** A distinct runtime instance built out of a class blueprint, allocated in the heap memory space[cite: 117]. [cite_start]Every Python object contains a hidden tracking field (`ob_refcnt`) that counts active references pointing to it[cite: 118].
* [cite_start]**C# Analogy:** An instantiated variable created via the `new` keyword[cite: 119].

---

- **5. del (and `__del__`)**
* [cite_start]**The Keyword (`del`):** Removes a variable name from the namespace and explicitly drops the target object's reference count by 1[cite: 120]. [cite_start]It does not directly clear out memory unless that specific decrement causes the counter to hit zero[cite: 121].
* [cite_start]**The Destructor (`__del__`):** The magic ("dunder") method called automatically by Python the exact moment an object's reference count drops to 0[cite: 122].
* [cite_start]**C# Analogy:** Acts similar to a C# Finalizer/Destructor (`~ClassName()`), but with one major advantage: it is highly deterministic for standard objects[cite: 123]. [cite_start]Rather than waiting for a periodic tracing cycle like the .NET CLR, Python reclaims the space and runs `__del__` instantly on the exact line where the last reference dies[cite: 124].

---

- Object Lifecycle Flow at a Glance

The flow below maps how these components work together inside Python's reference-counting engine:

1. [cite_start]**Instantiation:** `ClassSchedule('chemistry')` executes $\rightarrow$ Class allocates a new Object $\rightarrow$ `__init__` uses `self` to bind the Instance Variable $\rightarrow$ Reference count = 1[cite: 125].
2. [cite_start]**Execution:** Code references the object via its variable name[cite: 126].
3. [cite_start]**Reference Dropped:** You call `del` on the variable (or it naturally falls out of scope) $\rightarrow$ Reference count drops to 0[cite: 127].
4. [cite_start]**Destruction:** Python triggers the internal `__del__` destructor and instantly frees the memory[cite: 128].

````python
class ClassSchedule:
    def __init__(self,courseName):
        self.courseName = courseName

    def __del__(self):
        print(f"Delete is called for {self.courseName}", flush=True)    

objClassScheduleChemistry = ClassSchedule('chemistry')
print(objClassScheduleChemistry.courseName, flush=True)

objClassScheduleMaths = ClassSchedule('maths')
print(objClassScheduleMaths.courseName, flush=True)

del objClassScheduleChemistry

print("NOTE : Do not worry if your output order is wrong , its juypter notebook and vscode problem for correct output restart the notbook")        
````

# Access Modifications

## Public Access Modifier

By default, all members within a class are public, and there’s no need to specify access modifiers for public members. Being public means that these members can easily be accessed outside of the class

````python
class ClassSchedule:
    def __init__(self,course,instructor):
        self.course = course
        self.instructor = instructor

    def displayCourse(self):
        print(f"{self.course} -> {self.instructor}")

obj1 = ClassSchedule("checmistry","abcd")
obj1.displayCourse()
````

## Protected Access Modifier

Protected access modifiers, denoted with the prefix _, prevent members from being accessed outside of the class, unless it’s from a subclass. Let’s modify the class from the example above and make the members course and instructor protected with the _

````python
class ClassSchedule:
    def __init__(self,course,instructor):
        self._course = course
        self._instructor = instructor
 #The variables course and instructor are now protected members in the class. 
    def displayCourse(self):
        print(f"{self._course} -> {self._instructor}")

obj1 = ClassSchedule("checmistry","abcd")
obj1.displayCourse()
````

### Key Takeaway for a C# Developer

Protected Members in Python (_): In Python, prefixing an attribute with a single underscore _ (like _course) is purely a convention indicating that it is intended for internal/protected use.

Unlike C#, where the protected keyword is enforced at compile time, Python does not strictly block external access to single-underscore variables. However, following this naming convention signals to other developers that the variable should be treated as non-public.

# Encapsulation

In object-oriented programming (OOP), encapsulation is a fundamental concept that describes wrapping variables and methods in one unit. 
- A class can contain variables that can only be accessed by an object of the class. 
- A class can also contain methods that can be accessed by objects of the class. 

- In C#, encapsulation is enforced strictly by the compiler using access modifiers (public, private, protected). If you try to access a private field outside its class, C# will refuse to compile.

    In Python, encapsulation is centered around conventions, properties, and name mangling, operating under a famous core principle:

    **"We are all consenting adults here."**

    Python assumes developers understand the rules and won't deliberately break internal class state, so it doesn't build impassable compiler walls.

### 1. Access Modifiers: Conventions & Mechanics

Python does not enforce access modifiers in the same way that C# does. Instead, it relies on naming conventions and a mechanism called **name mangling** for attributes intended to be private.

The table below shows how access levels compare between C# and Python:

| Concept | C# Equivalent | Python Syntax | How Python Handles It |
|---------|---------------|---------------|------------------------|
| **Public** | `public string Course` | `self.course` | Fully accessible from anywhere. |
| **Protected** | `protected string Course` | `self._course` | A **convention only**. The single underscore (`_`) signals to developers that the attribute is intended for internal or subclass use. Python does **not** prevent external access. |
| **Private** | `private string Course` | `self.__course` | Uses **name mangling**. Python internally renames `__course` to `_ClassName__course` to reduce accidental access and avoid name collisions in subclasses. It is not true access control. |

> **Key takeaway:** Unlike C#, Python does not enforce access modifiers at the language level. Public, protected, and private access are implemented through conventions and name mangling rather than compiler-enforced restrictions.

````python
class UserInfo:
    def __init__(self,username,email):
        self.username = username
        self.email = email

    def checkUserName(self,usernameCheck):
        if usernameCheck == self.username:
            return True
        else:
            return False
        
obj1 = UserInfo('abcd','ddddd@email.com')

obj1.username
obj1.email

print(obj1.checkUserName('abcd'))
print(obj1.checkUserName('abc'))
````

## Private Access Modifier

Private access modifiers, denoted with the prefix __, declare members to be accessible within the class only. Members with this modifier will be marked private and any attempt to access these members outside of the class will cause an Attribute Error message. 

````python
class ClassSchedule:
   def __init__(self, course, instructor):
       self.__course = course # private
       self.__instructor = instructor # private
  
   def display_course(self):
       # public
 
       print(f'Course: {self.__course}, Instructor: {self.__instructor}')
 
sched = ClassSchedule('Biology', 'Ms. Smith')
 
#sched.__course # this will throw an Attribute Error because we're trying to access a private member
 
sched.display_course() # this won't throw an Attribute Error because this method is public
````

### In Python, private members are **"private by convention and design,"** not **"private by strict security or compiler enforcement."**

If you come from C#, this can feel strange. In C#, the compiler acts as a strict wall—if you try to access a private field outside the class, the code simply will not compile.

In Python, the design philosophy is famously summarized by the phrase:

> **"We are all consenting adults here."**

This philosophy means that developers are trusted to use class internals responsibly rather than being prevented from accessing them.

When you use double underscores (e.g., `__course`), here's what "private" actually means in Python:

#### 1. Name Mangling (How Python "Hides" It)

When you prefix an attribute name with double underscores, Python automatically renames it internally. This process is called **name mangling**.

For example, if you write:

```python
self.__course
```

inside the `ClassSchedule` class, Python internally renames it to:

```python
_ClassSchedule__course
```

Because of this renaming, attempting to access:

```python
sched.__course
```

from outside the class results in an `AttributeError`, because an attribute with that exact name does not exist.

#### 2. Why Python Still Allows Access

Although Python mangles the attribute name, it does **not** prevent access to it. If you explicitly use the mangled name:

```python
print(sched._ClassSchedule__course)
```

Python will happily print:

```text
Biology
```

#### Why Does Python Allow This?

Python's behavior is intentional.

- **Trust over restriction:** Python assumes developers are responsible and should only access private internals when there is a legitimate reason, such as debugging, testing, or serialization.
- **Name protection, not security:** Double underscores (`__`) are designed primarily to prevent accidental name collisions between parent and child classes during inheritance, **not** to provide true access control or data security.

> **Key takeaway:** In Python, double underscores provide **name mangling**, not true privacy. They discourage accidental access and help avoid inheritance conflicts, but they do not enforce strict access restrictions like `private` members in languages such as C#.

````python
class ClassSchedule:
    def __init__(self, course, instructor, code):
        self.course = course          # Public
        self._instructor = instructor  # Protected (convention)
        self.__code = code            # Private (mangled)

sched = ClassSchedule("Chemistry", "Dr. Stone", "CHEM101")

# Public
print(sched.course)       # Output: Chemistry

# Protected (Python allows it, but convention says don't)
print(sched._instructor)  # Output: Dr. Stone

# Private (Direct access fails)
# print(sched.__code)     # Raises AttributeError

# Private (Bypassing via Name Mangling)
print(sched._ClassSchedule__code)  # Output: CHEM101
````

## 2. Properties: The Pythonic Way (@property)

````python
class ClassSchedule:
    def __init__(self, course):
        self.course = course

    # Getter
    @property
    def course(self):
        print("getter")
        return self._course

    # Setter
    @course.setter
    def course(self, value):
        print("setter")
        if not value:
            print("[Warning] Empty value provided! Falling back to default.")
            self._course = "empty"
            #raise ValueError("Course name cannot be empty!")

        self._course = value   

obj1 = ClassSchedule("Math")
print(obj1.course)

obj1.course = "Chemistry"
print(obj1.course)

obj1.course = ""
print(obj1.course)
````

## Method Overriding

````python
class Person: #parent class
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def DisplayInfo(self):
        print(self.name)
        print(self.age)

class Teacher(Person): #child Class
    def __init__(self,name,age,subject):
        super().__init__(name,age)
        self.subject = subject
        
    def DisplayInfo(self):
        super().DisplayInfo()
        print(f"Override subject : {self.subject}")

        

obj1 = Teacher("Dr. Hirani", 49, "Computer Science")
obj1.DisplayInfo()
````

## Multiple Inheritance

````python
class Person: #parent class
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def DisplayInfo(self):
        print(self.name)
        print(self.age)

class Department:
    def __init__(self,deptName):
        self.deptName = deptName

    def DisplayInfo(self):
        print(f"Depratment class : {self.deptName}")
            

class Teacher(Person, Department): #child Class
    def __init__(self,name,age,deptName,subject):
        Person.__init__(self,name,age)
        Department.__init__(self,deptName)
        self.subject = subject
        

obj1 = Teacher("Dr. Hirani", 49, "CS department","Computer Science")
obj1.DisplayInfo()
Department.DisplayInfo(obj1)
````

## or  Composition (If you explicitly want obj1.department dot-notation)
If your goal is to use dot notation like obj1.department.DisplayInfo(), use Composition (has-a relationship) instead of Multiple Inheritance (is-a relationship):

````python
class Department:
    def __init__(self, deptName):
        self.deptName = deptName

    def DisplayInfo(self):
        print(f"Department class: {self.deptName}")

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Teacher(Person):
    def __init__(self, name, age, deptName, subject):
        super().__init__(name, age)
        # Store a Department instance as an attribute!
        self.department = Department(deptName)  
        self.subject = subject

obj1 = Teacher("Dr. Hirani", 49, "CS department", "Computer Science")

# 🟢 NOW THIS WORKS because self.department is an actual object attribute
obj1.department.DisplayInfo()
````

# Polymorphism

````python
class Dog:
    def speak(self):
        return "Woof!"

class Cat:
    def speak(self):
        return "Meow!"

class Robot:  # Completely unrelated class! No shared parent!
    def speak(self):
        return "Beep Boop!"

# Polymorphic function: Accepts ANY object that has a speak() method
def make_it_speak(self):
    print(self.speak())

# Execution
make_it_speak(Dog())    # Output: Woof!
make_it_speak(Cat())    # Output: Meow!
make_it_speak(Robot())  # Output: Beep Boop!
````

## Method Overriding

````python
# Parent Class (Base Class)
class Animal:
    def make_sound(self):
        print("Some generic animal sound")

# Child Class (Derived Class)
class Dog(Animal):
    # Overriding the parent's make_sound method
    def make_sound(self):
        print("Woof! Woof!")

obj1 = Animal()
obj2 = Dog()

obj1.make_sound()
obj2.make_sound()
````

# Method overloading and overriding in python

## 1. Method Overriding
    Method overriding occurs when a child class provides its own implementation of a method that is already defined in its parent class.

    Key Rules in Python:

    Virtual by Default: Unlike C# (where you must write public virtual void and public override void), all methods in Python are virtual by default.


    Same Signature: The child class defines a method with the exact same name as the parent class.


    super(): You can call the parent method inside the overridden method using super().method_name().

````python
class Animal:
    def make_sound(self):
        print("Some generic animal sound")

class Dog(Animal):
    # Overriding the parent's make_sound method
    def make_sound(self):
        super().make_sound()  # Calls parent method first
        print("Woof! Woof!")   # Adds Dog-specific behavior

# Execution
d = Dog()
d.make_sound()
````

## 2. Method Overloading
    Method overloading allows a class to have multiple methods with the same name but different signatures.

    🚨 The Python Trap: Native Overloading Does NOT Exist!
    If you write two methods with the same name in a Python class, Python does not throw a compile error—instead, the second definition completely overwrites the first one!
