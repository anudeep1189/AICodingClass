```python
#if

num = 80

print("###normal if ###")
if num > 40 :
   print(f"less than {num}")
else:
   print(f"not less than {num}")   

print("###nested if###")
if num < 40:
   print(f"{num} is less than 40")
   if num < 20:
    print(f"{num} is less than 20")
   else:
    print(f"{num} is more than 40")
else :
   print(f"{num} is more than 40")
   if num < 60:
    print(f"{num} is less than 60")
   else:
    print(f"{num} is greate than 40")
      

    
          
      
```

```python
# number is even or not

num = 8

if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

```python
#operation of datastructure 

num_list = [1,2,3,4,5,6]
num_tuple = (12,13,14,15)
num_set = {22,23,24,25}
num_dict = {'a':33,'b':34,'c':35,'d':36}

sum = 0 

for i in num_list:
    print(i)
    sum+=i
print(f"sum = {sum}") 

print("### dict ###")

for i,v in num_dict.items() :
    print(i,v)

    sum += v
print(f"sum = {sum}") 

print("------")
for i in num_dict.values() :
    print(i)

for i in num_dict.keys() :
    print(i)
```
