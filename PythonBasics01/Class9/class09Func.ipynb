```python
# list of dictonary data type
students = [
    {"name": "Aman", "scores": [80, 85, 90], "attendance": 92},
    {"name": "Riya", "scores": [95, 90, 93], "attendance": 96},
    {"name": "Karan", "scores": [55, 60, 58], "attendance": 70},
    {"name": "Neha", "scores": [72, 75, 78], "attendance": 88},
    {"name": "Mohit", "scores": [40, 45, 50], "attendance": 60},
]

```

```python

student_averages = list(map(
    lambda s: {"student_name": s["name"], "avg_marks": sum(s["scores"]) / len(s["scores"])}, 
    students
))

print(student_averages)
```
